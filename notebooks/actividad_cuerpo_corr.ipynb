```python
# -*- coding: utf-8 -*-
"""actividad_cuerpo_corr - Matriz de Correlación de Pearson
Señales fisiológicas: corazón, músculos y nivel de dolor (BioVid)
"""

# Montar Google Drive para acceder a los archivos del curso
from google.colab import drive
drive.mount('/content/drive')

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ─── 1. BUSCAR ARCHIVO ────────────────────────────────────────────────────────
RAICES = [
    '/content/drive/MyDrive',
    '/content/drive/Shareddrives',
    '/content/drive/Unidades compartidas', # La ruta puede variar según el idioma o configuración de Drive
]

for raiz in RAICES:
    if os.path.exists(raiz):
        print(f"\n--- Explorando: {raiz} ---")
        for root, dirs, files in os.walk(raiz):
            for file in files:
                if 'biovid' in file.lower():
                    print(os.path.join(root, file))

# ─── 2. CARGAR DATASET ───────────────────────────────────────────────────────
RUTA_ARCHIVO = '/content/drive/Shareddrives/DATA_GPI/Grupo2_Desarrollo/dataset_features_biovid.csv'

df_biovid = pd.read_csv(RUTA_ARCHIVO)

print(f"Filas: {df_biovid.shape[0]} | Columnas: {df_biovid.shape[1]}")
print("\nColumnas disponibles:")
print(df_biovid.columns.tolist())

# ─── 3. DEFINIR COLUMNAS POR CATEGORÍA (CORREGIDO) ───────────────────────────
# Columnas relacionadas con el nivel de dolor (variable objetivo)
cols_dolor = ['class_id']

# Columnas relacionadas con la señal cardíaca (ECG)
cols_corazon = ['ecg_bpm']

# NUEVO: Columnas relacionadas con la sudoración de la piel (GSR)
cols_sudor = [
    'gsr_peaks_count',
    'gsr_max_amplitude'
]

# Columnas relacionadas con las señales musculares (EMG)
cols_musculos = [
    'emg_trapezius_mean',
    'emg_corrugator_mean',
    'emg_zygomaticus_mean'
]

# Combinar TODAS las columnas de interés (Incluyendo sudor) para el análisis
COLUMNAS_INTERES = cols_dolor + cols_corazon + cols_sudor + cols_musculos

# ─── 4. VERIFICAR DISPONIBILIDAD ─────────────────────────────────────────────
columnas_faltantes = [col for col in COLUMNAS_INTERES if col not in df_biovid.columns]

if columnas_faltantes:
    raise ValueError(f"ERROR: Faltan las siguientes columnas en el dataset: {columnas_faltantes}")
else:
    print(f"\nOK: Todas las columnas requeridas están presentes.")

# ─── 5. FILTRAR Y LIMPIAR ────────────────────────────────────────────────────
# Seleccionar solo las columnas de interés y eliminar filas con valores nulos (NaN)
df_analisis = df_biovid[COLUMNAS_INTERES].dropna()

print(f"\nRegistros tras filtrar columnas y eliminar NaN: {df_analisis.shape[0]}")
print("Estadísticas descriptivas:")
print(df_analisis.describe().round(3))

# ─── 6. MATRIZ DE CORRELACIÓN DE PEARSON ─────────────────────────────────────
# Calcular la matriz de correlación de Pearson
corr_matrix = df_analisis.corr(method='pearson')

print("\nMatriz de correlación de Pearson:")
print(corr_matrix.round(3))

# ─── 7. VISUALIZACIÓN ────────────────────────────────────────────────────────
# Etiquetas legibles para los ejes del heatmap (Actualizado)
etiquetas = {
    'class_id':              'Nivel de Dolor',
    'ecg_bpm':               'FC (ECG bpm)',
    'gsr_peaks_count':       'Sudor (Picos GSR)',        # Agregado
    'gsr_max_amplitude':     'Sudor (Amp. Max GSR)',     # Agregado
    'emg_trapezius_mean':    'EMG Trapecio',
    'emg_corrugator_mean':   'EMG Corrugador',
    'emg_zygomaticus_mean':  'EMG Zigomático'
}

corr_plot = corr_matrix.rename(index=etiquetas, columns=etiquetas)

# Máscara para ocultar el triángulo superior del heatmap (evita duplicados y mejora la lectura)
mask = np.triu(np.ones_like(corr_plot, dtype=bool), k=1)

fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(
    corr_plot,
    mask=mask,
    annot=True,
    fmt='.3f',
    cmap='RdBu_r',        # 'RdBu_r': rojo para correlación positiva, azul para negativa
    vmin=-1, vmax=1,
    linewidths=0.5,
    linecolor='white',
    square=True,
    ax=ax,
    cbar_kws={'shrink': 0.8, 'label': 'r de Pearson'}
)

ax.set_title(
    'Correlación de Pearson\nCorazón · Músculos · Dolor (BioVid)',
    fontsize=13, fontweight='bold', pad=14
)
ax.tick_params(axis='x', rotation=30)
ax.tick_params(axis='y', rotation=0)

plt.tight_layout()
plt.savefig('correlacion_pearson_biovid.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nGráfico guardado como: correlacion_pearson_biovid.png")

# ─── 8. INTERPRETACIÓN AUTOMÁTICA ────────────────────────────────────────────
print("\n── Correlaciones con Nivel de Dolor (|r| > 0.1) ──")
corr_dolor = corr_matrix['class_id'].drop('class_id').sort_values(key=abs, ascending=False)

for col, r in corr_dolor.items():
    if abs(r) > 0.1: # Se muestran solo las correlaciones con una magnitud considerable.
        direccion = "positiva" if r > 0 else "negativa"
        intensidad = (
            "fuerte" if abs(r) > 0.5 else
            "moderada" if abs(r) > 0.3 else
            "débil"
        )
        print(f"  {etiquetas.get(col, col):25s} r = {r:+.3f}  ({intensidad}, {direccion})")
```
